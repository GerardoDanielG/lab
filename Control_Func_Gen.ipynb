```python
import pyvisa
from pyvisa import VisaIOError
```

```python
rm = pyvisa.ResourceManager()

for s in rm.list_resources():
    try:
        device = rm.open_resource(s)
        print(s, device.query('*IDN?').strip())
    except VisaIOError:
        continue
```

```python
rm = pyvisa.ResourceManager()
device = rm.open_resource('USB0::0x0699::0x034C::C010119::INSTR')  # where the FG is

print("Connected to:", device.query("*IDN?").strip())
```

```python
# to change amplitude

amp = 1.5 #Vpp

device.write(f'SOUR1:VOLT {amp}')

amp_actual = device.query('SOUR1:VOLT?')

print("Amplitude:", amp_actual)
```

```python
# to change frequency

freq = 100 #Hz

device.write(f'SOUR1:FREQ {freq}')

freq_actual = device.query("SOUR1:FREQ?")

print("Frequency:", freq_actual)
```

```python
# output of device (?)
out = device.query('OUTP1?')
print("Output:", out)
```
